```python
# Remove duplicate rows if any
df = df.drop_duplicates()
```

```python
# Fix rating_count column
# ----------------------------
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
# Fix price columns
# ----------------------------
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
# Fix discount percentage
# ----------------------------
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
```

```python
# Fix rating column
# ----------------------------
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

```python
# Create main category (optional)
# ----------------------------
df['main_category'] = df['category'].str.split('|').str[0]
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['main_category'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Categories by Number of Products")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['discounted_price'], y=df['rating'])
plt.title("Discounted Price vs Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['discount_percentage'], y=df['rating'])
plt.title("Discount Percentage vs Rating")
plt.xlabel("Discount %")
plt.ylabel("Rating")
plt.show()
```

```python
top_reviewed = df.sort_values(by='rating_count', ascending=False).head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_reviewed['rating_count'], y=top_reviewed['product_name'])
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Rating Count")
plt.ylabel("Product")
plt.show()
```

```python
avg_rating_cat = df.groupby('main_category')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
avg_rating_cat.plot(kind='bar')
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()
```

```python
# Save cleaned dataset for SQL / Power BI
df.to_csv("amazon_cleaned.csv", index=False)
```
